**------Q3------**

The difference between .fillna() and .dropna() is dropna() use to delete the rows and columns with missing values,whereas fillna() is used to fill those missing values with a specified value or method.

.fillna() is use when you want to preserve dataset size and the missing values can be reasonably estimated.

.dropna() is use when missing data is minimal or irrelevant, and removing it won’t harm representativeness.

```python
import pandas as pd
df=pd.read_csv('indian_startups_fundingDADS.csv')
#------cleaning------
# 1. Rename columns to clean snake_case
df.columns=(df.columns.str.strip().str.lower().str.replace('[^a-z0-9]+','_',regex=True).str.replace('_+','_',regex=True))

# 2. Clean startup_name and city
df['startup_name'] = df['startup_name'].str.strip().str.title()
df['city'] = df['city'].str.strip().str.title()

# 3. Standardise city aliases
city_map = {
    'Bangalore': 'Bengaluru', 'Blr': 'Bengaluru',
    'Bom': 'Mumbai', 'Bombay': 'Mumbai',
    'Ggn': 'Gurugram', 'Gurgaon': 'Gurugram',
    'Delhi': 'New Delhi',
    'Calcutta': 'Kolkata',
    'Madras': 'Chennai',
    'Hyd': 'Hyderabad'
}
df['city'] = df['city'].replace(city_map)

# 4. Standardise industry and funding_stage
industry_map = {
    'Ecommerce': 'E-Commerce', 'E-Commerce': 'E-Commerce',
    'Fintech': 'FinTech', 'Fin-Tech': 'FinTech',
    'Edtech': 'EdTech', 'Ed-Tech': 'EdTech',
    'Saas': 'SaaS', 'SaAs': 'SaaS',
    'Aiml': 'AI/ML', 'Ai Ml': 'AI/ML', 'Ai-Ml': 'AI/ML',
    'Proptech': 'PropTech', 'Prop-Tech': 'PropTech',
    'Foodtech': 'Food-Tech', 'Food Tech': 'Food-Tech',
    'Agritech': 'AgriTech', 'Agri Tech': 'AgriTech',
    'Beautytech': 'BeautyTech', 'Beauty-Tech': 'BeautyTech',
    'Logistics': 'Logistics',
    'Gaming': 'Gaming',
    'Ev': 'Electric Vehicles', 'Electric-Vehicles': 'Electric Vehicles'
}
stage_map = {
    'Seed Round': 'Seed', 'Seed': 'Seed',
    'Pre-Seed': 'Pre-Seed', 'Preseed': 'Pre-Seed',
    'Pre-A': 'Pre-A', 'Pre Series A': 'Pre-Series A', 'Pre-Series A': 'Pre-Series A',
    'Series A': 'Series A', 'Series-B': 'Series B', 'Series B': 'Series B',
    'Series C': 'Series C', 'Series D': 'Series D', 'Series E': 'Series E',
    'Bridge': 'Bridge',
    'Pre-Ipo': 'Pre-IPO', 'Pre-Ipo': 'Pre-IPO',
    'Ipo': 'IPO'
}
df['industry'] = df['industry'].str.strip().str.title().replace(industry_map)
df['funding_stage'] = df['funding_stage'].str.strip().str.title().replace(stage_map)

# 5. Convert amount to numeric (INR Cr)
df['amount'] = (df['amount'].astype(str).str.replace(r'[₹,]', '', regex=True).str.replace(r'INR|Cr|Crore', '', regex=True).str.strip().replace({'Undisclosed': None}))
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')

# 6. Convert funding_date to datetime
df['funding_date'] = pd.to_datetime(df['funding_date'], errors='coerce', dayfirst=True)

# 7. Drop duplicate rows
df = df.drop_duplicates()

# Final cleaned dataset
print(df.head(20))
df
```
